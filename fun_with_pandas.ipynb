## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
rawdata

# Check how many tables were extracted
print(f"Number of tables extracted: {len(rawdata)}")

# Inspect the first few tables to find the right ones
for idx, table in enumerate(rawdata):
    print(f"Table {idx}:")
    print(table.head(), "\n")  # Print the first few rows of each table
```

```python
p_singles = rawdata[0]
p_singles

p1_singles = rawdata[1]
p1_singles

p_single = pd.concat([p_singles, p1_singles])
p_single

p_single['Type'] = 'Physical'
p_single
```

```python
d_singles = rawdata[3]
d_singles

d1_singles = rawdata[4]
d1_singles

d_single = pd.concat([d_singles, d1_singles])
d_single

d_single['Type'] = 'Digital'
d_single
```

```python
combined = pd.concat([p_single, d_single])
combined
```

```python
#nothing for year 2000 so went with 1996 instead
born = []

# Loop through each row of the combined DataFrame
for idx, row in combined.iterrows():
    # Access the 'Released' year for the current row
    year = row['Released']
    
    # Check if the year is 1996
    if year == 1996:
        # Append a tuple of (Artist, Single) for rows where the year is 1996
        born.append((row['Artist'], row['Single']))

# Now the 'born' list contains tuples of (Artist, Single) for songs released in 1996
print(born)
```

```python
fifteen = []

# Loop through each row of the combined DataFrame
for idx, row in combined.iterrows():
    # Access the 'Released' year for the current row
    year = row['Released']
    
    # Check if the year is 2015
    if year == 2015:
        fifteen.append((row['Artist'], row['Single']))

# Now the 'fifteen' list contains tuples of (Artist, Single) 
fifteen
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
import pandas as pd

space_mission = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
space_mission
```

```python
#make all tables into normal tables
fifty = space_mission[0]
sixty = space_mission[1]
seventy = space_mission[2]
eighty = space_mission[3]
ninety = space_mission[4]
two = space_mission[5]
twoten = space_mission[6]
twotwenty = space_mission[7]

all_missions = pd.concat([fifty, sixty, seventy, eighty, ninety, two, twoten, twotwenty])
all_missions
```

```python
#converting the date to more readable format

all_missions['Launch date'] = pd.to_datetime(all_missions['Launch date'], format="%d %B %Y", errors='coerce', dayfirst=True)
all_missions
```

```python
#most launches in a year 

#extract only year 
all_missions['Year'] = all_missions['Launch date'].dt.year
year_counts = all_missions['Year'].value_counts()

# Find the year with the most launches
year_with_most_launches = year_counts.idxmax()
print(f"The year with the most launches is: {year_with_most_launches}")
```

```python
#most common month of launch

# Extract the month and count the occurrences
all_missions['Month'] = all_missions['Launch date'].dt.month
month_counts = all_missions['Month'].value_counts()

# Find the most common month for launches
most_common_month = month_counts.idxmax()
print(f"The most common month for launches is: {most_common_month}")
```

## Supervillain challenge

2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade

```python
supervillian = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
```

```python
#combine all tables

thirties = supervillian[3]
thirties['Decade'] = 30
forties = supervillian[4]
forties['Decade'] = 40
fifties = supervillian[5]
fifties['Decade'] = 50
sixties = supervillian[6]
sixties['Decade'] = 60
seventies = supervillian[7]
seventies['Decade'] = 70
eighties = supervillian[8]
eighties['Decade'] = 80
nineties = supervillian[9]
nineties['Decade'] = 90
twothousand = supervillian[10]
twothousand['Decade'] = 2000
twothouten = supervillian[11]
twothouten['Decade'] = 2010

all_villians = pd.concat([thirties, forties, fifties, sixties, seventies, eighties, nineties, twothousand, twothouten])
all_villians
```

```python
#decade with most villians

decade_count = all_villians['Decade'].value_counts()

# Find the year with the most launches
decade_most_villians = decade_count.idxmax()
print(f"The decade with most villian debuts is: {decade_most_villians}")
```

```python
company_count = all_villians['Company'].value_counts()

# Convert the Series to a DataFrame
company_count_df = company_count.reset_index()

# Rename the columns for clarity
company_count_df.columns = ['Company', 'Count']

company_count_df
```

## Stroop challenge

Every year between 2015 and 2021, the students in my Language, Cognition, and the Brain course participated in a version of the Stroop task. Using a stopwatch (ok, using their phones), they recorded how fast they could say a list of things (either reading or naming colors or color words). The column names mean "Reading with No Interference", "Naming with Interference", "Naming with No Interference", and "Reading with Interference". The times are in seconds.

### Stroop challenge 1: 
Transform these data from wide format to long format, so that the result is a dataframe with
- 1 column named "Participant_id" with a unique number for each participant (you can use the row indices)
- 1 column named "Year" with the year data
- 1 column named "Task" that shows which task they were doing
- 1 column named "RT" that shows their response time

```python
# Starter code 1...
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/Stroop/master/Stroop-raw-over-the-years.csv")
df.head()
```

```python
# Starter code 2...

# Make a new column using the dataframe indices as particpant numbers

#df.index.name = 'Participant_id'
#df = df.reset_index()

#df.reset_index(inplace = True)
# NOTE: This line does exactly the same thing as the line above:
# it replaces the original df with a new df with the updated index. That's what
# "inplace = True" means. Or, you can just assign the dataframe with the updated index
# to a new dataframe with the same name as the old dataframe, which is what I did above.
# The end result is the same.

df
```

```python
import pandas as pd

#Melt the DataFrame to long format
df['Participant_id'] = df.index

# Step 3: Melt the DataFrame to long format
long_df = df.melt(id_vars=['Participant_id', 'Year'],  
                  value_vars=['Reading_NoInt', 'Naming_Int', 'Naming_NoInt', 'Reading_Int'], 
                  var_name='Task', 
                  value_name='RT')

# Step 4: Optional - Rename Task for clarity (if needed)
long_df['Task'] = long_df['Task'].str.replace('_', ' ').str.replace('Int', ' Int').str.replace('NoInt', ' NoInt')

long_df = pd.DataFrame(long_df)

# Print the final long format DataFrame
print("\nLong DataFrame:")
print(long_df)
```

## Stroop challenge 2 (Advanced!!!):

Make a new dataframe which shows the mean response time (in seconds) for each task for each year.

```python
mean_rt_df = long_df.groupby(['Task', 'Year']).agg(mean_RT=('RT', 'mean')).reset_index()

# Step 6: Print the new DataFrame with mean response times
print("\nMean Response Time DataFrame:")
print(mean_rt_df)
```
